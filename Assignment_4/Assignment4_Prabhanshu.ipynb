```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('owid-co2-data-cleaned.csv')
df
```

```python
unique_countries = df['country'].unique()
print(f"Number of unique countries: {len(unique_countries)}")
print("Unique countries:", unique_countries)
```

```python
total_number_of_years = df['year'].nunique()
print(f"Total number of years in the dataset: {total_number_of_years}")
print("Years in the dataset:", df['year'].unique())
```

```python
print(f"Earliest year in the dataset: {df['year'].min()}")
print(f"Latest year in the dataset: {df['year'].max()}")
```

```python
df_latest_year = df[df['year']== df['year'].max()]
total_carbon_emissions = df_latest_year['co2'].sum()
print(f"Total carbon emissions in the latest year : {total_carbon_emissions} million tonnes")
```

```python
countries_with_highest_emissions = df_latest_year.nlargest(6, 'co2')
print(f"Countries with the highest carbon emissions in the latest year : {countries_with_highest_emissions[['country', 'co2']].set_index('country')}")
```

```python
co2_emmissions_by_india = df[df['country'] == 'India']
co2_india_each_year = co2_emmissions_by_india[['year', 'co2']].set_index('year')
co2_india_each_year
```

```python
maximum_co2_recorded = df['co2'].max()
minimum_co2_recorded = df['co2'].min()
print(f"Maximum CO2 recorded: {maximum_co2_recorded} million tonnes")
print(f"Minimum CO2 recorded: {minimum_co2_recorded} million tonnes")
```

```python
df_2010 = df[df['year'] == 2010]
lowest_emission_country_2010 = df_2010.loc[df_2010['co2'].idxmin()]
print(f"Country with the lowest CO2 emissions in 2010: {lowest_emission_country_2010['country']} with {lowest_emission_country_2010['co2']} million tonnes")
```

```python
print("Average CO2 emissions by country:")
print(df.groupby('country')['co2'].mean())
```

```python
df_sorted = df.sort_values(by='co2', ascending=False)
df_sorted
```

```python
total_emmissions_per_year = df.groupby('year')['co2'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=total_emmissions_per_year, x='year', y='co2')
plt.title('Total CO2 Emissions Over the Years')
plt.xlabel('Year')
plt.ylabel('Total CO2 Emissions (Million Tonnes)')
plt.yticks(np.arange(0,180000,10000))
plt.xticks(np.arange(1750, 2040, 15))
plt.grid(True)
plt.show()
```

```python
countries_with_highest_emissions = df_latest_year.nlargest(10, 'co2')
plt.figure(figsize=(25,8))
sns.barplot(data=countries_with_highest_emissions , x='country' , y='co2')
bars = plt.bar(countries_with_highest_emissions['country'], countries_with_highest_emissions['co2'])
plt.bar_label(bars)
plt.xlabel('Country')
plt.ylabel('CO2 emissions in 2023 (Million Tonnes)')
plt.title('Carbon Emissions of Top-10 Emitting countries in 2023')
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df['co2'])
plt.show()
```

```python
df_total_emission_by_country = df.groupby(['country'])['co2'].sum()
df_top5_country = df_total_emission_by_country.sort_values(ascending=False).head(5)
df_top5_country = pd.DataFrame(df_top5_country)
plt.pie(df_top5_country['co2'], labels=df_top5_country.index, autopct='%.2f%%')
plt.title('Top 5 CO2 emitting countries')
plt.show()
```

```python
df_india = df[df['country'] == 'India']
plt.figure(figsize=(15,10))
plt.plot(df_india['year'], df_india['co2'])
plt.xlabel('Year')
plt.ylabel('CO2 emissions (Million Tonnes)')
plt.title('India CO2 emissions over the year')
plt.show()
```

```python
df_2020 = df[df['year'] == 2020]
df_2020_ind = df_2020[df_2020['country']== 'India']
df_2020_china = df_2020[df_2020['country']== 'China']
df_2020_usa = df_2020[df_2020['country']== 'United States']
df_2020_usa_ind_china = pd.concat([df_2020_ind, df_2020_china, df_2020_usa])
```

```python
sns.barplot(data=df_2020_usa_ind_china, x='country', y='co2')
plt.xlabel('Country')
plt.ylabel('CO2 emissions in 2023 (Million Tonnes)')
plt.title('Bar Chart comparing CO2 emissions for China, USA, and India in 2020')
plt.show()
```
