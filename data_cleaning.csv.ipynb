```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('owid-co2-data.csv')
df
```

```python
threshold = 0.6  # 60% threshold for missing values
min_non_null_count = int((1 - threshold) * len(df))
df_cleaned = df.dropna(axis=1, thresh=min_non_null_count)
df_cleaned
```

```python
essential_columns = ['country', 'year', 'population', 'co2']
df_cleaned = df_cleaned.dropna(subset=essential_columns)
df_cleaned
```

```python
df_cleaned = df_cleaned.drop_duplicates()
df_cleaned
```

```python
df_cleaned.to_csv("owid-co2-data-cleaned.csv", index=False)
```
